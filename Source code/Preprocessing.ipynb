```python
import cv2
import os
import numpy as np

word_n    = 0
folder  = "DataCoba/UCF11_input"
folder1 = "DataCoba/UCF11_grayscale_noresize"
sample_n =  1

begin = 0
end = 1000

f = open("D:\\skripsi\labels.txt","r")
s = f.read()
f.close()
labels = s.split("\n")
label_n = len(labels)
jmlfile = 0
width = 224
height = 224
dim = (width, height)

while (True):
    if (word_n == label_n):
        break
    word = labels[word_n]
    path = folder + '/' + word + '/' + 'v' + '_' + word +  ' ' + '(' + str(jmlfile+1) + ')' + '.mp4'
    print (path)
    pathfolder = folder + '/' + word 
    files = [f for f in os.listdir(pathfolder) if os.path.isfile(os.path.join(pathfolder, f))]
    # Menghitung jumlah file
    file_count = len(files)
    print(f"Jumlah file dalam folder '{pathfolder}': {file_count}")
    vidcap = cv2.VideoCapture(path)
    def getFrame(sec):
        vidcap.set(cv2.CAP_PROP_POS_MSEC,sec*1000)
        ret,frame = vidcap.read()
        if ret:
            frame_path = folder1 + '/' + word + '-grayscale' + '/' + 'frame' 
            + word + '_' + str(jmlfile+1) +'_' + str(sample_n) + ".png"
            img = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
            cv2.imwrite(frame_path, resized)
            print("Save in : " + frame_path)
        return ret
    sec = 0
    frameRate = 0.15 #//it will capture image in each 0.15 second = 6 frame per detik
    count=1
    success = getFrame(sec)
    # print(success)
    while success:
        count = count + 1
        sample_n += 1
        sec = sec + frameRate
        sec = round(sec, 2)
        success = getFrame(sec)
        # print(success)
    if(success==False and sample_n==1 and jmlfile==file_count):
        print("Finished")
        break
    if(success==False and jmlfile==file_count):
        word_n += 1
        sample_n = 1
        jmlfile = 0
        word
        continue
    jmlfile = jmlfile + 1
```
